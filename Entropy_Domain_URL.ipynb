```python
import numpy as np
import pandas as pd
import time
from urllib.parse import urlparse
import warnings
warnings.filterwarnings('ignore')
URLs = pd.read_csv("dataset/TopWebs10M.csv",usecols=['Domain'])
urlCoun = len(URLs['Domain'])
urlCoun
```

```python
#Calculating each distinct character probability associated
x = [0] * 38
prob = [0] * 38
tot = 0
for i in range(urlCoun):
    url = URLs['Domain'][i]
    url = url.replace("https://","").replace("http://","").replace("www.","")
    for ch in url:
        ind = ord(ch)
        if(ind>=97 and ind<=122):
            ind = ord(ch)%97
        elif(ind>=48 and ind<=57):
            ind = 26 + ord(ch)%48
        elif(ind>=95):
            ind = 36
        else:
            ind = 37
        x[ind] = x[ind]+1
        tot = tot+1

tot = tot - x[37]
psum=0
for k in range(0,37):
    prob[k] = x[k]/tot
    psum = psum+prob[k]
    ch='_'
    if(k<=25):
        ch=chr(97+k)
    elif(k<=35):
        ch=chr(22+k)  #22=48-26
    print(ch," ", x[k], " ", str(round(prob[k], 4)))
print(psum)
```

```python
import math
URLs = pd.read_csv("dataset/P7_URLList_10May.csv")
urlCoun = len(URLs['url'])
SHAN = pd.DataFrame(columns=['FILENAME','url','domain'])
for i in range(urlCoun):
    url = URLs['url'][i]
    url = url.replace("https://","").replace("http://","")
    if(url.find('www.')==0):
        url = url.replace("www.","")
    urlLen = len(url)
    urlEntropy=0
    shannonEntropy=0
    for ch in url:
        ind = ord(ch)
        if(ind>=97 and ind<=122):
            ind = ord(ch)%97
        elif(ind>=48 and ind<=57):
            ind = 26 + ord(ch)%48
        elif(ind==95):
            ind = 36
        else:
            urlLen = urlLen-1
        if(ind<=36):
            urlEntropy = urlEntropy + prob[ind]
            shannonEntropy = shannonEntropy + prob[ind]*(math.log2(prob[ind])*(-1))
    newRow = {}
    newRow['FILENAME'] = URLs['FILENAME'][i]
    newRow['url'] = URLs['url'][i]
    newRow['domain'] = URLs['domain'][i]
    newRow['urlLen'] = urlLen
    newRow['proposedURLEntropy'] = urlEntropy
    newRow['proposedURLEntropyDiv'] = urlEntropy/urlLen
    newRow['shannonURLEntropy'] = shannonEntropy
    SHAN=SHAN.append(newRow , ignore_index=True)
    if(i%5000==0):
        print(i)
SHAN.to_csv('dataset/FeaturesMay10/P7_Entropy_URL_Char_May10.csv',index = False)
```
